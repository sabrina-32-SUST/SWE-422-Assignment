# Import libraries

```python

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline

```

```python
dataset = pd.read_csv('price_prediction.csv')
```

# Data Set

```python
dataset
```

```python
 sns.pairplot(dataset )
```

# CHECKING NULL

```python
dataset.isnull().any()
```

# Adding  Columns  into  Index

```python
dataset.columns
```

# Dividing Dataset in to x-array & y-array

```python
X = dataset[['bedrooms', 'bathrooms', 'sqft_living','sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'long', 'sqft_lot15']]
```

```python
X
```

```python
Y = dataset['price']
Y
```

# Splitting Dataset into Training Data and Test Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3)
```

# Building and Training Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

# Each modelâ€™s coefficients

```python
 model.coef_
```

# Intercept of the regression equation

```python
model.intercept_
```

```python
pd.DataFrame(model.coef_, X.columns, columns = ['Coeff'])
```

# Making Predictions From  Test Dataset

```python
predictions = model.predict(x_test)
```

# Scatterplot 

```python
plt.scatter(y_test, predictions)
```

# Visualization

```python
plt.hist(y_test - predictions)
```
